```python
N = 10000
L = 100

x = [0] * N
```

```python
import random
for i in range(N/10):
    start = random.randint(0, N - L)
    end = start + L
    for i in range(start, end):
        x[i] += 1
```

```python
plot(range(N), x)
axis(xmin=100, xmax=N-100)
title("Coverage by base")
xlabel("Position in genome")
ylabel("Number of times in data")
savefig('/tmp/cov.pdf')
```

```python
hist(x, bins=20)
title('Coverage distribution')
xlabel('Base coverage')
ylabel('N bases with that coverage')
savefig('/tmp/base.pdf')
```
